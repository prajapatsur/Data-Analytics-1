#Data Transformation

```python
df["Month"]=df["Date"].dt.month_name()   #Converted the date into months

#Now the task is to add this "Month" just next to the Date column.
#I can do that by extracting the index of "Date" and then adding the "Month" at +1 index.
date_index = df.columns.get_loc('Date')
# Insert the "Month" column next to the "Date" column with the data
df.insert(date_index + 1, 'Month', df.pop('Month'))
```

```python
nomi=pgeocode.Nominatim('in')
```

```python
# Convert to list and clean data
postal_codes = df['ship-postal-code'].astype(str).str.strip().tolist()

# Query postal codes and handle errors
try:
    results = nomi.query_postal_code(postal_codes)
    # Extract state names
    state_names = results['state_name']
except ValueError as e:
    print(f"Error querying postal codes: {e}")
    # Handle the error appropriately
```

```python
state_names
```

```python
postal_code_index = df.columns.get_loc('ship-postal-code')
df["State-name"] = state_names
```

This step showing some error, so we got an indication that there is some problem with value counts. So we will check for Nan in this new column "State-name".

```python
state_names.isna().sum()
```

We found Nan values in this column "State-name". So we will first set all these Nan values to "Unknown". Then will add the "State-name" column.

```python
df.pop('State-name')
```

```python
state_names = state_names.replace(np.nan, "Unknown")
state_names.isna().sum()
```

```python
df["State-name"] = state_names
df.insert(postal_code_index+1, "State-name", df.pop('State-name'))
```

```python
df.info()
```

```python
def age_category(x):
  if x<=10:
    return "0-10"
  elif x>10 and x<=20:
    return "10s"
  elif x>20 and x<=30:
    return "20s"
  elif x>30 and x<=40:
    return "30s"
  elif x>40 and x<=50:
    return "40s"
  elif x>50 and x<=60:
    return "50s"
  elif x>60 and x<=70:
    return "60s"
  elif x>70 and x<=80:
    return "70s"
  else:
    return "80+"
```

```python
df["Age-Category"] = df["Age"].apply(age_category)
age_index = df.columns.get_loc('Age')
df.insert(age_index+1, "Age-Category", df.pop("Age-Category"))
```

#Data Analytics

```python
import matplotlib.pyplot as plt
```

```python
#male-female pie chart
gender_distribution= df["Gender"].value_counts()
print(gender_distribution);
plt.figure(figsize=(2,2))
plt.pie(
    gender_distribution.values,
    labels=["Male", "Female"],
    shadow=True,
    colors=["Red","Pink"]
)
plt.title('Distribution of Men and Women')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
#Order Status pie chart
order_status = df['Status'].value_counts()
print(order_status);

def func(pct, allvalues):
    absolute = int(pct / 100. * sum(allvalues))
    return f'({pct:.1f}%)'

explode_values = [0.4] * len(order_status)


plt.figure(figsize=(4,4))
plt.pie(order_status,
        labels=order_status.index,
        autopct=lambda pct: func(pct, gender_distribution.values),
        explode= explode_values)
plt.show();
```

```python
age_category_counts = df["Age-Category"].value_counts()
print(age_category_counts);
plt.figure(figsize=(4,3))
plt.bar(age_category_counts.index, age_category_counts.values)
plt.xlabel("Age Category")
plt.ylabel("Number of Customers")
plt.title("Distribution of Customers Across Age Categories")
plt.show();
```

```python
# Pivot table
pivot_table = pd.pivot_table(df, index="Month", values=["Order ID", "Amount"], aggfunc={"Order ID": "count", "Amount": "sum"})
pivot_table = pivot_table.rename(columns={"Order ID": "Count of Order ID", "Amount": "Sum of Amount"})
pivot_table
```
